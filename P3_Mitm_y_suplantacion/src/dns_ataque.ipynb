```python
from scapy.all import *
import random
import time


DNS_SERVER = "10.0.0.53"

def random_subdomain():
    return ''.join(random.choice("abcdefghijklmnopqrstuvwxyz") for _ in range(10)) + ".test.local"

print("[*] Iniciando ataque de DNS snooping (simulado)...")

for i in range(100):

    domain = random_subdomain()

    pkt = IP(dst=DNS_SERVER)/UDP(dport=53)/DNS(rd=1, qd=DNSQR(qname=domain))

    send(pkt, verbose=0)

    print(f"[DNS QUERY] {domain}")

    time.sleep(0.1)
```
